```python
from pathlib import Path
import json

systems = ['6h86_A', '6p5h_A', '6jv8_A', '7lp1_A', '7jfl_C']
models = ['ProtSCAPE', 'MDGen', 'AlphaFLOW']
metrics_dir = Path("/home/lcl35/project_pi_sk2433/lcl35/ProtSCAPE-Net/experiments")

metrics_list = [
    ("rmwd_metrics", "rmwd"),
    ("rmwd_metrics", "translation"),
    ("rmwd_metrics", "variance"),
    ("rmsf_corr_metrics", "rmsf_corr"),
    ("pool_pca_metrics", "w2"),
    ("ref_pca_metrics", "w2"),
    ("pairwise_rmsd_metrics", "percent_diff")
]
metric_names = [f"{'_'.join(c.split('_')[:-1])}_{m}" for c, m in metrics_list]
```

```python
import pandas as pd

data = []
for model in models:

    for system in systems:


        for trial in range(1, 6):
        
            metrics_path = metrics_dir / f"{model}_metrics" / f"ATLAS_1K_trial_{trial}" / system / f"{system}_metrics.json"

            with open(metrics_path, 'r') as f:
                metrics_data = json.load(f)
            
            entry = [model, system, trial]
            for category, metric_name in metrics_list:
                value = metrics_data[category][metric_name]
                entry.append(value)
            data.append(entry)

df_metrics = pd.DataFrame(data, columns=['model', 'system', 'trial'] + metric_names)
df_metrics
```

```python
for model in models:

    for system in systems:
        df_system = df_metrics[(df_metrics['system'] == system) & (df_metrics['model'] == model)]

        output = []
        print(f"System: {system}, Model: {model}")

        for metric in metric_names:
            values = df_system[metric].values
            mean = values.mean()
            std = values.std()
            output.append(f"{metric}: {mean:.4f} ± {std:.3f}")
        print("\n".join(output))
```
